```python
import pandas as pd
df=pd.read_csv('indian_election.csv')
```

```python
df.info()
```

```python
df.rename(columns={'st_name':'State','cand_name':'Candidate_name','cand_sex':'Gender','totvotpoll':'total_vote_received','electors':'No._of_voters'},inplace=True)
```

```python
df.info()
```

```python
df.drop(columns=['partyabbre','Candidate_name','pc_name','pc_no'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()  
```

```python
df.rename(columns={'pc_type':'Constituency_reservation_status'},inplace=True)
```

```python
df.info()
```

```python
df.dropna(subset=['Gender'],inplace=True)
```

```python
df.info()
```

```python
df['State']=df['State'].astype('string')
df['year']=df['year'].astype('string')
df['Constituency_reservation_status']=df['Constituency_reservation_status'].astype('string')
df['Gender']=df['Gender'].astype('string')
df['partyname']=df['partyname'].astype('string')
```

```python
df.info()
```

```python
df['Constituency_reservation_status'].value_counts()
```

```python
df['Constituency_reservation_status']=df['Constituency_reservation_status'].fillna('General')
```

```python
df.info()
```

```python
#saving data
df.to_csv('Indian_election_cleandata.csv',index=False)
```
